# Group By

Group By is a function which aggregates the data and groups it by the requiered operation 

```python
#The reason we use "as_index=False" is because we don't want the 'Rep' to be our index. 
#Let us see what will happen if we remove that part. 

df.groupby('Rep').sum()
```

```python
df.groupby('Rep',as_index=False).sum() 
```

```python
#Now, we need to only see the Total section, hence: 
df.groupby('Rep', as_index=False)['Total'].sum()
```

```python
#Using the agg function is much more convinenent 
df.groupby('Rep', as_index=False)['Total'].agg('mean')
```

```python
#Using the agg function is much more convinenent 
df.groupby('Rep', as_index=False)['Total'].agg(['sum','mean','max'])
```

```python
df.groupby('Rep')[['Total', 'Unit Cost']].agg({'Total':['max', 'min'], 'Unit Cost':['mean', 'sum']})
```

```python
df.groupby(['Rep','Region']).agg('sum')
```
