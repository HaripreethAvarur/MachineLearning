```python
import pandas as pd
import numpy as np
```

```python
data_df=pd.read_csv("C:/Users/Haripreeth Avarur/Desktop/Python/Machine Learning/CombinedCyclePowerPlant_UCI.csv")
```

```python
data_df.head()
```

```python
#Independent Variables: AT, V, AP, RH
#Dependent Variable: PE
```

```python
#x=independent variables 
x=data_df.drop(['PE'],axis=1).values
#y=dependent variable
y=data_df['PE'].values
```

```python
x
```

```python
y
```

#### Split the datasheet into training and test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

#### Training the model

```python
from sklearn.linear_model import LinearRegression
ml=LinearRegression()
ml.fit(x_train,y_train)
```

#### Predicting the Result

```python
y_pred=ml.predict(x_test)
y_pred
```

```python
ml.predict([[8.34,40.77,1010.84,90.01]])
```

#### Evaluate the model 

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

#### Plot

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(25,20))
plt.scatter(y_test,y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
```

#### Predicted Values

```python
pred_y_df=pd.DataFrame({'Actual Value':y_test,'Predicted Value':y_pred, 'Difference':y_test-y_pred})
pred_y_df#[0:20]
```
