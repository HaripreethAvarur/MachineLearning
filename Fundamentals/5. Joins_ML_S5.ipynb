Types of Joins:

    i. Inner 
    ii. Left 
    iii. Rigt 
    iv. Full 

Note: Here, Left Table means base table or 1 of the tables, and Right Table is the auxiallary table. This can be understood better by viewing this in a venn diagram as seen below. Your final table should be the left table. Because the left table is the table that will get modified to become the final table.

Inner Join:

    If I have 2 tables. This function will output only the common rows.
    
Left Join:

    If I have 2 tables. This function will output the common rows and everything on the left table.
    
Right Join: 

    If I have 2 tables. This function will output the common rows and everything on the right table.
    
Full Join: 

    If I have 2 tables. This function will output everything from both the tables.
    


##### Inner Join

```python
innerJoin = emp.merge(dep, left_on = 'Dept_Id', right_on = 'ID', how = 'inner')
#innerJoin is a new table name; the base table must come in the left, so that is 'dep')
innerJoin


#Here we can see that '10' from 'Dept_Id' and '5' from 'ID' is not there as it's not a common row.
```

##### Left Join

```python
leftJoin = emp.merge(dep, left_on = 'Dept_Id', right_on = 'ID', how = 'left')
leftJoin
```

##### Right Join 

```python
rightJoin = emp.merge(dep, left_on = 'Dept_Id', right_on = 'ID', how = 'right')
rightJoin
```

##### Full Join

```python
fullJoin = emp.merge(dep, left_on = 'Dept_Id', right_on = 'ID', how = 'outer')
fullJoin

#rememeber the function here is called 'outer'
```
