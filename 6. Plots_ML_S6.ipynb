```python
import matplotlib.pyplot as plt 
```

##### Scatter Plot: 

```python
# X-axis Values:
x = [5, 2, 9, 4, 7]

# Y-axis Values:
y = [10, 5, 8, 4, 2]

# Function to plot a scatter graph:
plt.scatter(x,y)

# Function to show the plot:
plt.show()
```

##### Line Chart:

```python
# X-axis Values:
x = [5, 2, 9, 4, 7]

# Y-axis Values:
y = [10, 5, 8, 4, 2]

# Function to plot a scatter graph:
plt.plot(x,y)

# Function to show the plot:
plt.show()
```

##### Bar Graph:

```python
# X-axis Values:
x = [5, 2, 9, 4, 7]

# Y-axis Values:
y = [10, 5, 8, 4, 2]

# Function to plot a scatter graph:
plt.bar(x,y)

# Function to show the plot:
plt.show()
```

##### Histogram: 

```python
# Y-axis Values:
y = [1, 2, 3, 4, 5, 1, 2, 1, 6, 7, 8, 9, 10]

# Function to plot a histogram:
plt.hist(y)

# Function to show the plot:
plt.show()
```

```python
x = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]
num_bins = 5 #no. of bins is basically the no. of bars
plt.hist(x, num_bins, facecolor = 'orange')
plt.show()
```
