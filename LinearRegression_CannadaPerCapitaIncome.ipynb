```python
X=df.drop("Per_Capita_Income_USD",axis=1) #X will store Year as Per_Capita_Income_USD is dropped 
```

```python
y=df.drop("Year",axis=1) #y will store Per_Capita_Income_USD as Year is dropped 
```

```python
X=np.asarray(X) #This is converted into a 2D Array
y=np.asarray(y) #This is converted into a 2D Array
```

```python
#Building a linear regression model
lm = linear_model.LinearRegression()
#Fitting the data in the model
model = lm.fit(X, y)
model
```

```python
#Prediction
pred_year=np.asarray([[2020]])
pred=model.predict(pred_year)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(X,y)
```

```python
#Visualize the result
df.plot(kind='scatter', x='Year', y='Per_Capita_Income_USD')
#Plotting the regression line
plt.plot(X,model.predict(X),color='orange', linewidth=2)
#plotting the predicted value 
plt.scatter(pred_year,pred[[0]], color='green')
#plt.plot(x,y,color='blue',linewidth=1)
plt.show()
```
